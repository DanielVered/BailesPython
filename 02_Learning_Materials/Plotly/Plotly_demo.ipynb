```python
canada_data = px.data.gapminder().query("country=='Canada'")
fig = px.line(canada_data , x="year", y="lifeExp", title='Life expectancy in Canada' )
fig.show()
```

```python
oceania_data = px.data.gapminder().query("continent == 'Oceania'")
fig = px.line(oceania_data, x='year', y='lifeExp', color='country')
fig.show()
```

```python
canada_data = px.data.gapminder().query("country == 'Canada'")
px.bar(canada_data, x='year', y='pop')
```

```python
bills_data = px.data.tips()
fig = px.histogram(bills_data, x="total_bill")
fig.show()
```

```python
fig = px.scatter(x=[0, 1, 2, 3, 4], y=[0, 1, 4, 9, 16])
fig.show()
```

```python
px.scatter(bills_data ,x="total_bill", y="tip", trendline="ols")
```

```python
fig = px.box(bills_data, x="time", y="total_bill")
fig.show()
```
