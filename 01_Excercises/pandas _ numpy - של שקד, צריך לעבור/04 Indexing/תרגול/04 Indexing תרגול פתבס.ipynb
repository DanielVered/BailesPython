<div dir = "rtl">
### ברשומות שהאינדקס שלהן הוא 'versicolor' החלף את הערכים בעמודה 'petal_width' בערך Nan. 

```python
iris.loc['versicolor', 'petal_width'] = np.nan
```

<div dir = "rtl">
### בעזרת הפונקציה fillna(), החלף את כל ערכי ה-nan בערך 5.0. 

```python
iris.petal_length.fillna(5.0, inplace = True)
iris
```

<div dir = "rtl">
### החזר את האינדקס המקורי (לרסט את האינדקס). 

```python
iris.reset_index(inplace = True)
```

<div dir = "rtl">
### מחק את עמודת 'class'. 

```python
del iris['class']
iris.head()
```

<div dir = "rtl">
### קבע את כל ערכי 5 הרשומות הראשונות ל- Nan. 

```python
iris.iloc[0:5 ,:] = np.nan
iris.head()
```

<div dir = "rtl">
### מחק את כל הרשומות שישה בהם את הערך NaN. 

```python
iris = iris.dropna(how='any')
iris.head()
```
