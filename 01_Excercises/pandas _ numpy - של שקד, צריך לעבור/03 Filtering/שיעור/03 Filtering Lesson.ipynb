## Filter with More than One Condition (AND)

```python
df = pd.read_csv("https://raw.githubusercontent.com/kita8/PythonCsvData/main/employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"

df[mask1 & mask2]
```

## Filter with More than One Condition (OR)

```python
df = pd.read_csv("https://raw.githubusercontent.com/kita8/PythonCsvData/main/employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask1 = df["Senior Management"]
mask2 = df["Start Date"] < "1990-01-01"

df[mask1 | mask2]
```

```python
mask1 = df["First Name"] == "Robert"
mask2 = df["Team"] == "Client Services"
mask3 = df["Start Date"] > "2016-06-01"

df[(mask1 & mask2) | mask3]
```

```python
mask1 = df["First Name"] == "Robert"
mask2 = df["Team"] == "Client Services"
mask3 = df["Start Date"] > "2016-06-01"

df[(mask1 & mask2) | mask3]
```

## The `.isin()` Method

```python
df = pd.read_csv("https://raw.githubusercontent.com/kita8/PythonCsvData/main/employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask1 = df["Team"] == "Legal"
mask2 = df["Team"] == "Sales"
mask3 = df["Team"] == "Product"

df[mask1 | mask2 | mask3]
```

```python
mask = df["Team"].isin(["Legal","Product","Sales"])
```

```python
df[mask]
```

## The `.duplicated()` Method

```python
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace = True)
df.head(3)
```

```python
mask = ~df["First Name"].duplicated(keep = False) # Drop all duplicated values, even the first one
df[mask]
```

## The `np.where()` Method

```python
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask = df['Start Date'] > '1993-01-01'
```

```python
np.where(mask,'Yes','No')
```

```python
len(np.where(mask,'Yes','No'))
```

```python
len(df)
```

```python
df['Is Senior?'] = np.where(mask,'Yes','No')
```
