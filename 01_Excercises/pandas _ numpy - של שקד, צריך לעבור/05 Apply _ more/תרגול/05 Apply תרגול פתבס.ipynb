# Student Alcohol Consumption

<div dir = "rtl">
### קראו את הקובץ 'Alcohol.csv' לתוך DF בשם 'alcohol'.

```python
alcohol = pd.read_csv("https://raw.githubusercontent.com/kita8/PythonCsvData/main/Alcohol.csv")
alcohol.head()
```

<div dir = "rtl">
### הדפס כל עמודה בנפרד.

```python
alcohol.apply(lambda x : print(x))
```

<div dir = "rtl">
### הוסף לכל ערך בעמודה G3 את המספר 10. עשה זאת בלי Apply:

```python
alcohol["G3"] =alcohol["G3"]+10 
```

<div dir = "rtl">
### הוסף לכל ערך בעמודה G3 את המספר 10. עשה זאת עם Apply:

```python
alcohol["G3"] =alcohol["G3"].apply(lambda x : x+10)
```

<div dir = "rtl">
### הפוך כל מילה לאותיות גדולות בעמודה Fjob.

```python
alcohol["Fjob"] =alcohol["Fjob"].apply(lambda x : x.upper())
```

<div dir = "rtl">
### צור עמודה חדשה שנקראית adult. אם הגיל שווה או גדול מ-18 אז הערך יהיה True ואחרת יהיה False.

```python
alcohol["adult"] =alcohol["age"].apply(lambda x : True if x >= 18 else False)
```

<div dir = "rtl">
### הדפס כל רשומה בנפרד.

```python
alcohol.apply(lambda x : print(x), axis =1)
```

<div dir = "rtl">
### צור עמודה חדשה שנקראית 'organization' - עבור כל רשומה - אם סכום הערכים G1, G2,G3 גדולים מ- 35 אז הערך יהיה : "mossad", אחרת הוא יהיה: "shabak".

```python
def sum_row(row):
    if sum([int(row["G1"]), int(row["G2"]), int(row["G3"])]) > 35:
        return "mossad"
    return "shabak"
alcohol["organization"] = alcohol.apply(sum_row, axis =1)
```

<div dir = "rtl">
### בנה פונקציה שמדפיסה את 30 הרשומות הראשונות ב-DF באופן הבא:
<div dir = "ltr">
### This age is : (age)
### The school is : (school)
### His total Grades are : (G1 + G2 + G3)
### He's from : (organization)


```python
def print_row(row):
    sentence = """
    This age is : {}
    The school is : {}
    His total Grades are : {}
    He's from : {}
    """.format(row["age"], row["school"], sum([int(row["G1"]), int(row["G2"]), int(row["G3"])]), row["organization"])
    print(sentence)
    
alcohol.iloc[:30].apply(print_row, axis =1)
```

<div dir = "rtl">
### בונוס: הכפל כל מספר ב-DF ב-10.

```python
def mul_10(row):
    for i in range(len(row)):
        try :
             row[i] = int(row[i]) * 10
        except Exception:
            pass
    return row
```

```python
alcohol.apply(mul_10, axis=1)
```
