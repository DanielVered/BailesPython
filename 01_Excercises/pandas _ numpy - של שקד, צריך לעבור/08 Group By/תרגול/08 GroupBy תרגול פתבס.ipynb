# Regiment

<div dir = "rtl">
### לפניכם מילון. הכניסו את הנתונים בו ל-DF בשם regiment.

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
```

```python
regiment = pd.DataFrame(raw_data)
regiment
```

<div dir = "rtl">
### מה ממוצע לפני ואחרי המבחן עבור כל גדוד (regiment)?

```python
regiment.groupby('regiment').mean()
```

<div dir = "rtl">
### מה ממוצע לפני ואחרי המבחן עבור גדוד ה-Nightwaks?

```python
regiment[regiment['regiment'] == 'Nighthawks'].groupby('regiment').mean()
```

<div dir = "rtl">
### הצג נתונים סטטיסטים כללים עבור כל חברה (company). מהי הפונקציה המתאימה?

```python
regiment.groupby('company').describe()
```

<div dir = "rtl">
### הציגו את ממוצע הערכים ב-DF בחלוקה לחברה וגדוד.

```python
regiment.groupby(['regiment', 'company']).mean()
```

<div dir = "rtl">
### מה ממוצע הציון לפני המבחן עבור כל חברה? וודאו שהנתונים מוצגים כ-SERIES.

```python
regiment.groupby('company').preTestScore.mean()
```

<div dir = "rtl">
### מה ממוצע הציון אחרי המבחן עבור כל חברה וגדוד? וודאו שהנתונים מוצגים כ-SERIRS.

```python
regiment.groupby(['regiment', 'company']).postTestScore.mean()
```

<div dir = "rtl">
### מה ממוצע הציון אחרי המבחן עבור כל חברה וגדוד? הפכו את ה-SERIES ל-DF בעל אינדקס יחיד.

```python
regiment.groupby(['regiment', 'company']).preTestScore.mean().unstack()
```

```python
regiment.groupby(['regiment', 'company']).preTestScore.mean().to_frame().unstack()
```

<div dir = "rtl">
### עבור כל חברה וגדוד, כמה ערכים קיימים?

```python
regiment.groupby(['company', 'regiment']).size()
```

<div dir = "rtl">
### בצעו לולאה עבור הנתונים בחלוקה לגדודים. הדפיסו בכל ריצה של הלולאה את שם הגדוד ואת הקבוצה עצמה (ה-DF).

```python
# Group the dataframe by regiment, and for each regiment,
for name, group in regiment.groupby('regiment'):
    # print the name of the regiment
    print(name)
    # print the data of that regiment
    print(group)
```

# Titanic

<div dir = "rtl">
### קראו את הקובץ 'titanic.csv' לתוך DF בשם 'df'.

```python
df = pd.read_csv("https://raw.githubusercontent.com/kita8/PythonCsvData/main/titanic_data.csv")
df.head()
```

<div dir = "rtl">
### כמה גברים וכמה נשים היו על הספינה? חשבו איזה שדה לחשב.

```python
df.groupby("Sex").nunique()["PassengerId"]
```

<div dir = "rtl">
### חשבו את הציון הממוצע עבור כל מין.

```python
df.groupby("Sex").mean()["Age"]
```

<div dir = "rtl">
### השדה Survived מכיל 1 אם הנוסע שרד את הנסיעה ו-0 אם הוא מת :( חשבו את אחוז ההישרדות בקרב כל הנוסעים.

```python
df["Survived"].mean()
```

<div dir = "rtl">
### חשבו את אחוז ההישרדות בקרב נוסעים שקטנים מ-25.

```python
df.query("Age < 25")["Survived"].mean()
```

<div dir = "rtl">
### מי שרד יותר? נשים או גברים? חשבו את אחוז ההישרדות של כל מין.

```python
df.groupby("Sex").mean()["Survived"]
```

<div dir = "rtl">
### השדה Cabin מורכב מאות ומספר, כאשר האות מייצגת את האזור של התא. 
### אם הייתי רוצה לשרוד בטיטאניק. על איזו מחלקה (Class) ואזור תא (אות ראשונה של Cabin) הייתם ממליצים?

```python
df["Cabin_Area"] = df["Cabin"].str.slice(0,1)
```

```python
df.groupby(["Pclass", "Cabin_Area"]).mean()["Survived"].sort_values(ascending = False)
```

# Alcohol

<div dir = "rtl">
### קראו את הקובץ 'Alcohol_Consumption.csv' לתוך DF בשם 'drinks'.

```python
drinks = pd.read_csv('https://raw.githubusercontent.com/kita8/PythonCsvData/main/Alcohol.csv')
drinks.head()
```

<div dir = "rtl">
### באיה יבשת שותים הכי הרבה בירה?

```python
drinks.groupby('continent').beer_servings.sum()
```

<div dir = "rtl">
### עבור כל יבשת, הציגו נתונים סטטיסטים על צריכת היין.

```python
drinks.groupby('continent').wine_servings.describe()
```

<div dir = "rtl">
### עבור כל יבשת, הציגו ממוצע של כל השדות.

```python
drinks.groupby('continent').mean()
```

<div dir = "rtl">
### עבור כל יבשת, הציגו חציון של כל השדות.

```python
drinks.groupby('continent').median()
```

<div dir = "rtl">
### מעצבן לעשות הכל בפונקציה נפרדת. עבור כל יבשת, בצעו את הפעולות החישוביות הבאות בפונקציה יחידה: מינימום, מקסימום, חציון וממוצע.  

```python
drinks.groupby('continent').spirit_servings.agg(['mean', 'median', 'min', 'max'])
```
