## The `.agg()` Method

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby("Sector")
fortune.head(3)
```

<div class = "alert alert-block alert-info">
The agg method allows us a bit more flexibility when it comes to running aggregative functions on grouped data. <br>
agg receives a dictionary where the keys are the columns and the values are the methods - let's see some examples.

```python
sectors.agg({"Revenue" :"sum",
             "Profits" : "sum",
              "Employees" : "mean"})
```

```python
sectors.agg({"Revenue" :["sum", "min", "max"],
             "Profits" : ["sum", "mean"],
              "Employees" : "mean"})
```

```python
sectors.agg(["size", "sum", "mean"]) #We can give just lists
```

## Iterating through Groups

<div class = "alert alert-block alert-info">
Remember when we tried finding the highest grossing company in every sector? We got this:

```python
sectors["Revenue"].max()
```

<div class = "alert alert-block alert-info">
But this doesn't give us all the data... what we need is to iterate through the groups

```python
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for sector, data in sectors: #We have 2 variables because python will automatically use the first to reference the group object, and the second will be the actual data frame
    highest_revenue_company_in_group = data.nlargest(1, "Revenue") #nlargest extracts the n(1 this time) largest values - this time we specified it to the Revenues column
    df = df.append(highest_revenue_company_in_group)
```

```python
df
```

<div class = "alert alert-block alert-info">
Technically another way to get this result is .first(), because this DF is sorted by revenue... <br>
Let's try the same thing with a different groupby object

```python
cities = fortune.groupby("Location")
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for city, data in cities:
    highest_revenue_in_city = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_in_city)
```
